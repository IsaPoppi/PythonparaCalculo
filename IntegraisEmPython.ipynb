Calculando Integrais e Teorema de green em Python


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, diff, Eq, solve
from sympy.vector import CoordSys3D

N = CoordSys3D('N')

# Defina as variÃ¡veis simbÃ³licas
x, y = symbols('x y')

# Defina as funÃ§Ãµes para P e Q (componentes do campo vetorial)
P = x**2 + y
Q = -x

# Defina a curva fechada (exemplo: um cÃ­rculo)
x_curve = x**2 + y**2 - 4  # EquaÃ§Ã£o do cÃ­rculo x^2 + y^2 = 4

# Encontre a equaÃ§Ã£o da curva em termos de uma variÃ¡vel (por exemplo, x)
y_curve = solve(Eq(x_curve, 0), y)

# Calcule a integral de linha de Pdx + Qdy ao longo da curva fechada
line_integral = integrate(P.diff(x).subs(y, y_curve[0]) + Q.diff(y).subs(x, x_curve), (x, -2, 2))

# Calcule a integral dupla da divergÃªncia de (Q, -P) sobre a regiÃ£o limitada pela curva
divergence = (Q.diff(x) + P.diff(y)).simplify()
surface_integral = integrate(divergence, (x, -2, 2), (y, -2, 2))

# Verifique se o Teorema de Green Ã© satisfeito (a integral de linha Ã© igual Ã  integral de superfÃ­cie)
if line_integral == surface_integral:
    print("O Teorema de Green Ã© satisfeito.")
else:
    print("O Teorema de Green nÃ£o Ã© satisfeito.")

# Crie uma figura para plotar a curva e o campo vetorial
fig, ax = plt.subplots()

# Plote a curva C (cÃ­rculo)
theta = np.linspace(0, 2 * np.pi, 100)
x_values = 2 * np.cos(theta)
y_values = 2 * np.sin(theta)
ax.plot(x_values, y_values, label='Curva C (CÃ­rculo)')

# Plote o campo vetorial (P, Q)
x_range = np.linspace(-2, 2, 20)
y_range = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x_range, y_range)
U = X**2 + Y
V = -X
ax.quiver(X, Y, U, V, scale=15, color='r', label='Campo Vetorial (P, Q)')

ax.set_aspect('equal', adjustable='box')
ax.legend()
plt.title('Curva C e Campo Vetorial')
plt.grid()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

from sympy import symbols, integrate, diff, Eq, solve

from sympy.vector import CoordSys3D

N = CoordSys3D('N')

# Defina as variÃ¡veis simbÃ³licas

x, y = symbols('x y')

# Defina as funÃ§Ãµes para P e Q (componentes do campo vetorial)

P = x**2

Q = y**2

# Defina a curva fechada (uma elipse)

# EquaÃ§Ã£o da elipse: (x^2/4) + (y^2/1) = 1

x_curve = x**2/4 + y**2/1 - 1

# Encontre a equaÃ§Ã£o da curva em termos de uma variÃ¡vel (por exemplo, x)

y_curve = solve(Eq(x_curve, 0), y)
# Calcule a integral de linha de Pdx + Qdy ao longo da curva fechada

line_integral = integrate(P.diff(x).subs(y, y_curve[0]) - Q.diff(y).subs(x, x_curve), (x, -2,
2))

# Calcule a integral dupla da divergÃªncia de (Q, -P) sobre a regiÃ£o limitada pela curva

divergence = (Q.diff(x) + P.diff(y)).simplify()

surface_integral = integrate(divergence, (x, -2, 2), (y, -1, 1))

# Crie uma figura para plotar a curva e o campo vetorial

fig, ax = plt.subplots()

# Plote a curva C (elipse)

theta = np.linspace(0, 2*np.pi, 100)

x_values = 2 * np.cos(theta)

y_values = 1 * np.sin(theta)

ax.plot(x_values, y_values, label='Curva C (Elipse)')
# Plote o campo vetorial (P, Q)

x_range = np.linspace(-2, 2, 20)

y_range = np.linspace(-1, 1, 20)

X, Y = np.meshgrid(x_range, y_range)

U = X**2

V = Y**2

ax.quiver(X, Y, U, V, scale=10, color='r', label='Campo Vetorial (P, Q)')

ax.set_aspect('equal', adjustable='box')

ax.legend()

plt.title('Curva C (Elipse) e Campo Vetorial (P, Q)')

plt.grid()

plt.show()

print("Integral de Linha:", line_integral)

print("Integral de SuperfÃ­cie:", surface_integral)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Defina a curva C (uma parte do perÃ­metro do quadrado)
C_x = [0, 1, 1, 0, 0]
C_y = [0, 0, 1, 1, 0]

# Defina o quadrado
square_x = [0, 1, 1, 0, 0]
square_y = [0, 0, 1, 1, 0]

# FunÃ§Ã£o para calcular a integral de linha
def line_integral(x, y):
    return x**3 + y**3 - 2*x*y

integral = 0

for i in range(4):
    delta_x = C_x[i + 1] - C_x[i]
    delta_y = C_y[i + 1] - C_y[i]
    integral += line_integral(C_x[i], C_y[i]) * delta_x + line_integral(C_x[i], C_y[i]) * delta_y

# Plote a curva C com sentido anti-horÃ¡rio
plt.plot(C_x, C_y, label='Curva C (Sentido Anti-horÃ¡rio)', color='blue')

# Plote o quadrado
plt.plot(square_x, square_y, linestyle='--', label='Quadrado', color='red')

# Destaque a equaÃ§Ã£o da curva
plt.text(0.5, 0.5, r'$x=0, y=0, x=1, y=1$', fontsize=12, color='blue')

# Destaque a integral de linha
plt.text(0.5, 0.1, f'Integral de Linha: {integral:.2f}', fontsize=12, color='green')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Curva C e Quadrado com Nova Integral de Linha')
plt.legend()
plt.grid(True)
plt.show()
```
