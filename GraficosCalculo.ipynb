```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def função(x,y):
  return 3*x**2 + 2*y**2 + 7*y+10

# Crie uma série de valores de x
x = np.linspace(-10, 10, 100) # por exemplo, de -10 a 10 com 100 pontos
y = np .linspace(-20,20,100)
x, y = np.meshgrid(x, y)
# Calcule os valores correspondentes de função(x)
z= função(x,y)

# Plote o gráfico
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('função(x, y)')
ax.set_title('Gráfico da função')
plt.show()
```

# Gradiente descendente

```python
import numpy as np
import matplotlib.pyplot as plt
# Função de exemplo: f(x) = x^2
def função_de_exemplo(x):
 return x**2 +6

# Derivada da função de exemplo: f'(x) = 2x
def derivada_da_função(x):
  return 2 * x

 # Configurações do gradiente descendente
taxa_aprendizado = 0.1
n_iterações = 2000
x_inicial = 4 # Valor inicial arbitrário

# Listas para armazenar os valores de x e f(x) durante as iterações
valores_x = [x_inicial]
valores_f = [função_de_exemplo(x_inicial)]




# Algoritmo de gradiente descendente
for _ in range(n_iterações):
  gradiente = derivada_da_função(x_inicial)
  x_inicial = x_inicial - taxa_aprendizado * gradiente
  valores_x.append(x_inicial)
  valores_f.append(função_de_exemplo(x_inicial))

 # Plotagem do gráfico
plt.figure(figsize=(10, 5))
plt.plot(valores_x, valores_f, marker='o', linestyle='-', color='b')
plt.title('Descida de Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
