# **O que é uma Equação Diferencial Ordinária (EDO)?**

É uma matemática que descreve como alguma coisa muda com o tempo. Por exemplo, pode ser usado para descrever como a população de uma cidade cresce ou como a temperatura muda durante o dia.É como uma fórmula que nos ajuda a entender essas mudanças.


---


**Como o programa resolve EDOs?**
Aqui está uma explicação passo a passo de como o programa resolve EDOs:

1. **Entrada da EDO:** Primeiro, você fornece ao programa a EDO que descreve a mudança que deseja entender. Por exemplo, você pode dizer que a mudança é representada pela equação "0.1 * y - 2 * x". Isso significa que a taxa de mudança da quantidade "y" é afetada por essa fórmula, onde "x" é o tempo.

2. **Conversão Simbólica:** O programa converte essa equação em algo que ele possa entender matematicamente.

3. **Condições Iniciais:** Você também precisa informar ao programa em que ponto no tempo deseja começar a calcular. Por exemplo, você pode dizer que, no início, "x" é 0 e "y" é 1.

4. **Método de Euler:** O programa usa o Método de Euler, uma técnica matemática, para calcular como a quantidade "y" muda em pequenos intervalos de tempo. Ele faz isso dividindo o tempo em pedaços menores, chamados "passos", e calcula uma estimativa de "y" para cada um desses passos.

5. **Gráfico da Solução:** O programa mostra um gráfico que ilustra como a quantidade "y" muda com o tempo. No eixo horizontal (à direita), você verá o tempo (ou "x" na equação), e no eixo vertical (acima), você verá o valor de "y". O gráfico mostra uma curva que representa essa mudança ao longo do tempo.

6. **Resultado Numérico:** Abaixo da equação que você inseriu, o programa também fornece um número. Esse número é uma estimativa do valor de "y" em um ponto específico no tempo, de acordo com a EDO e os cálculos do programa.

No final, o programa ajuda a visualizar e entender como a quantidade "y" muda com o tempo, tanto graficamente quanto numericamente, usando a equação diferencial que você forneceu. Isso é útil para muitas áreas, desde a física até a biologia, para entender o comportamento de sistemas que mudam com o tempo.


# **OBS:**
Quanto menor o tamanho do passo, mais precisa será a aproximação. Se você aumentar o `tamanho_passo`, terá menos números quebrados, mas a solução será menos precisa.

O tamanho do passo é uma escolha importante ao resolver equações diferenciais numericamente. Pequenos passos fornecem uma solução mais precisa, mas também aumentam a quantidade de cálculos necessários. Grandes passos podem ser mais rápidos, mas a solução será menos precisa. Portanto, o tamanho do passo é um compromisso entre precisão e eficiência computacional.

```python

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Passo 1: Introdução e instruções
print("Bem-vindo ao resolvedor de Equações Diferenciais Ordinárias (EDOs)!")
print("Vamos resolver uma EDO simples usando o Método de Euler.")
print("Uma EDO descreve como algo muda com o tempo.")

# Passo 2: Explicação da EDO
print("Primeiro, vamos definir a EDO. Ela segue o formato padrão: dy/dx = f(x, y).")
print("Você deve nos fornecer a equação f(x, y) que descreve o comportamento da função y.")

# Passo 3: Solicitar a EDO ao usuário
equacao_diferencial = input("Insira a equação f(x, y) no formato Python (exemplo: 0.1 * y - 2 * x): ")

# Passo 4: Definir as variáveis simbólicas
x, y = sp.symbols('x y')

# Passo 5: Converter a entrada do usuário em uma expressão simbólica
f = sp.sympify(equacao_diferencial)

# Passo 6: Solicitar condições iniciais ao usuário
print("Agora, precisamos de condições iniciais para resolver a EDO.")
x0 = float(input("Insira o valor inicial de x: "))
y0 = float(input("Insira o valor inicial de y: "))

# Passo 7: Explicação do Método de Euler
print("Ótimo! Agora, vamos configurar o Método de Euler para encontrar a solução numérica.")
print("O Método de Euler divide o intervalo de x em pequenos passos e calcula a solução aproximada.")

# Passo 8: Definir o tamanho do passo e o número de pontos
h = float(input("Insira o tamanho do passo de integração (por exemplo, 0.1): "))
n = int(input("Insira o número de pontos (por exemplo, 100): "))

# Passo 9: Inicializar listas para armazenar os valores de x e y
x_values = [x0]
y_values = [y0]

# Passo 10: Converter a EDO em uma função Python
slope_function = sp.lambdify((x, y), f, 'numpy')

# Passo 11: Aplicar o Método de Euler para encontrar a solução
for i in range(n):
    x_new = x_values[-1] + h  # Calcular o próximo valor de x
    y_new = y_values[-1] + h * slope_function(x_values[-1], y_values[-1])  # Calcular o próximo valor de y
    x_values.append(x_new)
    y_values.append(y_new)

# Passo 12: Exibir a solução em um gráfico
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_values, y_values, label='Solução numérica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Solução da Equação Diferencial')

# Passo 13: Exibir a equação diferencial
plt.subplot(1, 2, 2)
plt.axis('off')
plt.text(0.1, 0.8, f'EDO: dy/dx = {equacao_diferencial}', fontsize=12)
plt.text(0.1, 0.4, 'Solução Numérica:', fontsize=12)
plt.text(0.1, 0.2, f'y({x_values[-1]:.2f}) = {y_values[-1]:.4f}', fontsize=12)
plt.title('Equação Diferencial e Solução Numérica')

plt.tight_layout()
plt.show()

# Passo 14: Fim
print("A solução da EDO foi calculada e exibida em um gráfico, e a equação diferencial também foi mostrada, junto com o valor numérico da solução.")
```

# **ENTENDENDO A SOLUÇÃO**

Suponha que você tenha uma equação que descreva como algo muda com o tempo, por exemplo, "dy/dx = 0.1 * y - 2 * x". Essa equação representa uma situação hipotética, onde "y" é uma quantidade que está mudando, e essa mudança é governada pela equação.

O programa usa o Método de Euler para calcular uma estimativa dessa mudança ao longo do tempo. Ele faz isso dividindo o tempo em pequenos passos e calculando aproximadamente o valor de "y" em cada passo. Por exemplo, começa com um valor inicial de "y" e, em seguida, calcula o valor de "y" no próximo momento, depois no próximo, e assim por diante.

Quando o programa termina de calcular, ele exibe dois resultados:

1. **O Gráfico:** Um gráfico mostra como o valor de "y" muda ao longo do tempo. No eixo horizontal (x), você verá o tempo (ou "x" no contexto da equação), e no eixo vertical (y), você verá o valor de "y". O gráfico mostra uma curva que representa como "y" muda conforme o tempo avança.

2. **A Solução Numérica:** Logo abaixo da equação diferencial que você inseriu (por exemplo, "dy/dx = 0.1 * y - 2 * x"), o programa mostra um valor numérico. Isso é como o valor de "y" é aproximadamente em um ponto específico do tempo. Por exemplo, ele pode dizer algo como "y(2.00) = 3.45". Isso significa que quando o tempo é 2.00 (ou "x" é 2.00), a estimativa numérica para "y" é 3.45. É uma forma de entender o valor de "y" em um momento específico de acordo com a equação e o Método de Euler.

Em resumo, o programa usa a equação que você forneceu para estimar como uma quantidade ("y") muda com o tempo. Ele mostra essa mudança em um gráfico e também fornece um valor numérico para essa quantidade em um momento específico do tempo. Isso ajuda a entender como o fenômeno descrito pela equação se comporta ao longo do tempo.
